```r
mothurtre <- read.newick(file = "/content/final.opti_mcc.jclass.0.03.tre")

normalize <- function(x, na.rm = TRUE) {
  return((x - min(x)) / (max(x) - min(x)))
}

mothurtre$edge.length <- normalize(mothurtre$edge.length - 0.5)

label_new <- c(mothurtre$tip.label, rep("", 18))

ggtree(mothurtre)
```

```r
ggtree(mothurtre) +
  geom_text(
    label = label_new,
    size = 1,
    hjust = -0.2,
    color = "firebrick"
  )
ggsave("Phylo200.png", width = 10, height = 10, dpi = 300)
```

```r
ggtree(mothurtre, ladderize = FALSE) +
  geom_text(
    label = label_new,
    size = 1,
    hjust = -0.2,
    color = "firebrick"
  )
#ggsave("Phylo200.png", width = 10, height = 10, dpi = 300)
```
