```r
library(ggtree)
library(ggplot2)
library(treeio)
```

```r
mothurtre <- read.newick(file = "/content/final.opti_mcc.thetayc.0.03.tre")

normalize <- function(x, na.rm = TRUE) {
  return((x - min(x)) / (max(x) - min(x)))
}

mothurtre$edge.length <- normalize(mothurtre$edge.length - 0.5)

label_new <- c(mothurtre$tip.label, rep("", 18))

ggtree(mothurtre)
```

```r
ggtree(mothurtre) +
  geom_text(
    label = label_new,
    size = 4,
    hjust = -0.2,
    color = "firebrick"
  )
```

```r
ggtree(mothurtre, ladderize = FALSE) +
  geom_text(
    label = label_new,
    size = 4,
    hjust = -0.2,
    color = "firebrick"
  )
```

```r
ggtree(mothurtre,
       layout = "circular",
       alpha = 0.9,
       size = 3) +
  geom_tiplab(aes(angle = angle), color = 'blue')

label_manual <- seq(1,37)
```

```r
ggtree(mothurtre,
       layout = "circular",
       alpha = 0.9,
       size = 1) +
  #geom_tiplab(aes(angle = angle), color = 'blue') +
  geom_text(
    aes(label = label_manual),
    size = 6,
    vjust = 0,
    color = "firebrick"
  )
```

```r
ggtree(mothurtre,
       layout = "fan",
       alpha = 0.9,
       size = 1) +
  geom_text(
    aes(label = label_new),
    size = 3,
    vjust = 0,
    color = "red"
  ) +
  geom_cladelab(
    node = 27,
    label = "This is a clade",
    align = TRUE,
    offset = 0.2,
    geom = 'label',
    fill = 'lightblue'
  )
```

```r
ggtree(mothurtre,
       layout = "fan",
       alpha = 0.9,
       size = 1) +
  geom_text(
    aes(label = label_new),
    size = 3,
    vjust = 0,
    color = "red"
  ) +
  geom_cladelab(
    node = 5,
    label = "This is a clade",
    align = TRUE,
    offset = 0.2,
    geom = 'label',
    fill = 'lightblue'
  ) +
  geom_hilight(node=30,
               fill="red")+
  geom_hilight(node=21,
               fill="green")
```
